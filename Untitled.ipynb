```python
quotient = 1
rows = quotient = quotient + 1
columns = 3
```

```python
a = {}
a[1] = 1.1
a[2] = 1.9
a[3] = 0.7
a[4] = 1.5
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(columns*20, rows*15))
for i in range(3):
    plt.subplot(rows, columns, i+1)
    plt.plot(list(a.keys()), list(a.values()), linewidth=5)
plt.savefig('saved_files/test.pdf', format='pdf', dpi=1200)
plt.show()
```
